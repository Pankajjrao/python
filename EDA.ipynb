```python
# creating converter function to convert the columns form object type to float which are in thousands
```

```python
def converter(df):     # Takes an dataframe as an argumnet
    for i in df:        # iterating over the columns of the dataframe
        if i in ['reviews','salaries','interviews','jobs','benefits']: # seleting the cokumns which are needed to converted
            df[i] = df[i].str.replace('k','').astype(float).astype(int)*1000 
    # we take each column as a series and replace the 'k' in it with '' 
    # then we change the datatype to float because it cannot be convertted directly into int 
    # because it conatains decimal type
    # then it is converted into int at last which is required
    return df

converter(df)
    
```
